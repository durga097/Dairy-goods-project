# is total land area depends on number of cows

```python
number_of_cows = np.array(df["Total Land Area (acres)"])
farm_size = np.array(df["Number of Cows"])
my_rho = np.corrcoef([number_of_cows, farm_size])
print(my_rho)
```

```python
#with the help of uppper matrix we can say that tatal land area not depends on number of cows
```

# mention the name of the products with max and min shelf life


```python
high_shelf_life = df["Shelf Life (days)"].idxmax()
high_shelf_life
```

```python
low_shelf_life = df["Shelf Life (days)"].idxmin()
low_shelf_life
```

# summarize the data on the basis of brands and land acquired by them,cows and the revenue generated by them. 


```python
group_by = df.groupby(['Brand']).agg({"Total Land Area (acres)":"sum","Number of Cows":"sum", "Approx. Total Revenue(INR)":"sum"})
group_by
```
